```python
import os

# Organizing File
date = '2024-12-05'
main_path = f'source/{date}'

# Loop through files in the main directory
for file in os.listdir(main_path):
    if file.endswith('.avi'):
        # Extract timestamp from original filename
        # Original format: tram drop off_192.168.100.105_20241202_073839_826171217.avi
        parts = file.split('_')
        timestamp = parts[3]  # Gets the timestamp (073839)
        
        # Format the new filename
        formatted_date = date.replace('-', '')  # 20241202
        new_filename = f'ayana_tram_stop_{formatted_date}_{timestamp}.avi'
        
        # Rename the file
        old_path = f'{main_path}/{file}'
        new_path = f'{main_path}/{new_filename}'
        
        os.rename(old_path, new_path)
        print(f"Renamed: {file} -> {new_filename}")

print("Files have been renamed successfully!")
```
